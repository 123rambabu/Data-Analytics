```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('HR_Analytics.csv', encoding= 'unicode_escape')
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'AgeGroup',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs Monthly Income

sales_gen = df.groupby(['Gender'], as_index=False)['MonthlyIncome'].sum().sort_values(by='MonthlyIncome', ascending=False)

sns.barplot(x = 'Gender',y= 'MonthlyIncome' ,data = sales_gen)
```

```python
sales_state = df.groupby(['BusinessTravel'], as_index=False)['EnvironmentSatisfaction'].sum().sort_values(by='EnvironmentSatisfaction', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'BusinessTravel',y= 'EnvironmentSatisfaction')
```

```python
ax = sns.countplot(data = df, x = 'AgeGroup', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Age vs MonthlyRate

sales_age = df.groupby(['Age'], as_index=False)['MonthlyRate'].sum().sort_values(by='MonthlyRate', ascending=False)

sns.barplot(x = 'Age',y= 'MonthlyRate' ,data = sales_age)
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['JobRole'], as_index=False)['MonthlyRate'].sum().sort_values(by='MonthlyRate', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'JobRole',y= 'MonthlyRate')
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['JobRole'], as_index=False)['MonthlyIncome'].sum().sort_values(by='MonthlyIncome', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'JobRole',y= 'MonthlyIncome')
```

```python
sales_state = df.groupby(['AgeGroup', 'Gender'], as_index=False)['PercentSalaryHike'].sum().sort_values(by='PercentSalaryHike', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'AgeGroup',y= 'PercentSalaryHike', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'PercentSalaryHike')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['EnvironmentSatisfaction'], as_index=False)['EmployeeCount'].sum().sort_values(by='EmployeeCount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'EnvironmentSatisfaction',y= 'EmployeeCount')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'YearsAtCompany')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['YearsAtCompany'], as_index=False)['Age'].sum().sort_values(by='Age', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'YearsAtCompany',y= 'Age')
```

```python
sales_state = df.groupby(['WorkLifeBalance'], as_index=False)['DailyRate'].sum().sort_values(by='DailyRate', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'WorkLifeBalance',y= 'DailyRate')
```
